## The case of [20 20 20]

```julia
evolEE0s = Float64[]
evolEE4s = Float64[]
evolEE6s = Float64[]
evolEE8s = Float64[]
evolEE10s = Float64[]

io = open("/Users/tianyiyan/Desktop/Chaos/data/ees/ee/text0.txt", "r") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(evolEE0s, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/ees/ee/text4.txt", "r") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(evolEE4s, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/ees/ee/text6.txt", "r") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(evolEE6s, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/ees/ee/text8.txt", "r") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(evolEE8s, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/ees/ee/text10.txt", "r") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(evolEE10s, ln)
    end
end
```

```julia
fig = Figure()
lines(fig[1,1], ts, evolEE0s,linewidth=1.5)
lines!(fig[1,1], ts, evolEE4s, linewidth=1.5)
lines!(fig[1,1], ts, evolEE6s, linewidth=1.5)
lines!(fig[1,1], ts, evolEE8s, linewidth=1.5)
lines!(fig[1,1], ts, evolEE10s, linewidth=1.5, color=:black)
fig
```
