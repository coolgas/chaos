## Read the average values of ee and std

```julia
ave = Float64[]
ave_new = Float64[]
ave_newnew = Float64[]

io = open("/Users/tianyiyan/Desktop/Chaos/data/aves/avee/aveEEs.txt") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(ave, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/aves/avee_new/aveEEs_new.txt") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(ave_new, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/aves/avee_newnew/aveEEs_newnew.txt") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(ave_newnew, ln)
    end
end

std = Float64[]
std_new = Float64[]
std_newnew = Float64[]

io = open("/Users/tianyiyan/Desktop/Chaos/data/aves/log_stdee/text0.txt") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(std, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/aves/log_stdee_new/text0.txt") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(std_new, ln)
    end
end

io = open("/Users/tianyiyan/Desktop/Chaos/data/aves/log_stdee_newnew/text0.txt") do io
    for ln in eachline(io)
        ln = parse(Float64, ln)
        append!(std_newnew, ln)
    end
end
```

```julia
fontsize_theme = Theme(fontsize = 25)
set_theme!(fontsize_theme)

f = Figure(resolution=(1000,1000))
ax1 = Axis(f[1,1], ylabel=L"S_{EE}")
lines!(ax1, ts, evolEE0s, label=L"$\gamma=0$")
lines!(ax1, ts, evolEE4s, label=L"$\gamma=4")
lines!(ax1, ts, evolEE6s, label=L"$\gamma=6$")
lines!(ax1, ts, evolEE8s, label=L"$\gamma=8")
lines!(ax1, ts, evolEE10s, color="red", label=L"$\gamma=10$")
hidexdecorations!(ax1,grid=false)
Label(f[1, 1], "(a)", tellwidth=false, tellheight=false,valign = :top, halign = :left, padding = (280, 5, 5, 3))


ax2 = Axis(f[2,1], ylabel=L"S_{EE}")
lines!(ax2, ts, evolEE0s_new)
lines!(ax2, ts, evolEE4s_new)
lines!(ax2, ts, evolEE6s_new)
lines!(ax2, ts, evolEE8s_new)
lines!(ax2, ts, evolEE10s_new, color="red")
hidexdecorations!(ax2,grid=false)
Label(f[2, 1], "(b)", tellwidth=false, tellheight=false,valign = :top, halign = :left, padding = (280, 5, 5, 3))

ax3 = Axis(f[3,1], xlabel=L"$t$", ylabel=L"S_{EE}")
lines!(ax3, ts, evolEE0s_newnew)
lines!(ax3, ts, evolEE4s_newnew)
lines!(ax3, ts, evolEE6s_newnew)
lines!(ax3, ts, evolEE8s_newnew)
lines!(ax3, ts, evolEE10s_newnew, color="red")
Label(f[3, 1], "(c)", tellwidth=false, tellheight=false,valign = :top, halign = :left, padding = (280, 5, 5, 3))

ax4 = Axis(f[1,2], ylabel=L"\log(\bar{S}_{EE})")
scatter!(ax4, gammas_new, log.(ave), color="black")
errorbars!(gammas_new, log.(ave), std, whiskerwidth=10)
hidexdecorations!(ax4,grid=false)
Label(f[1, 2], "(d)", tellwidth=false, tellheight=false,valign = :top, halign = :left, padding = (280, 5, 5, 3))

ax5 = Axis(f[2,2], ylabel=L"\log(\bar{S}_{EE})")
scatter!(ax5, gammas_new, log.(ave_new), color="black")
errorbars!(gammas_new, log.(ave_new), std_new, whiskerwidth=10)
hidexdecorations!(ax5,grid=false)
Label(f[2, 2], "(e)", tellwidth=false, tellheight=false,valign = :top, halign = :left, padding = (280, 5, 5, 3))

ax6 = Axis(f[3,2], xlabel=L"$\gamma$", ylabel=L"\log(\bar{S}_{EE})")
scatter!(ax6, gammas_new, log.(ave_newnew),  color="black")
errorbars!(gammas_new, log.(ave_newnew), std_newnew, whiskerwidth=10)
Label(f[3, 2], "(f)", tellwidth=false, tellheight=false,valign = :top, halign = :left, padding = (280, 5, 5, 3))

f[:,3] = Legend(f, ax1, framevisible=false)

f
```
