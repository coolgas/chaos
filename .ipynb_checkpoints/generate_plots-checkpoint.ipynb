## Number evolutions

```julia
using Chaos
using LinearAlgebra
```

```julia
Na = 60; # number of atoms
Ls = 3; # number of wells

Hbs = integer_partition(N=Na, L=Ls);
Lb = length(Hbs) # the dimension of the Hilbert space
```

```julia
HLs = zeros(Float64, Lb, Lb)
for i in 1:Ls-1
    HLs += hopping_pair(m=i, Na=Na, Hbs=Hbs)
end
Hhop = HLs + HLs';

Honsite = zeros(Float64, Lb, Lb);
for j in 1:Lb
    for k in 1:Lb
        if j == k
            sum = 0
            for l in 1:Ls
                sum += Hbs[j][l]^2 - Hbs[j][l]
            end
            Honsite[j,k] += sum
        end
    end
end

function Htit(::Float64)
    Htit = zeros(Float64, Lb, Lb)
    for j in 1:Lb
        for k in 1:Lb
            if j == k
                for l in 1:Ls
                    Htit[j,k] += -(l-(Ls+1)/2)**Hbs[j][l]
                end
            end

        end
    end
    return Htit
end

(delta::Int64, d::Float64, C6::Float64, R::Float64) = C6 / ((delta)^6*d^6 + R^6)

# Here 'd' is the lattice constant.
function Hsc(d::Float64, C6::Float64, R::Float64)
    Hsc = zeros(Float64, Lb, Lb)
    for j in 1:Lb
        for k in 1:Lb
            if j == k
                for l in 1:Ls
                    for m in 1:Ls
                        if l-m == 0
                            Hsc[j,k] += (0,d,C6,R)*Hbs[j][l]*Hbs[j][m]
                        elseif abs(m-l) == 1
                            Hsc[j,k] += (1,d,C6,R)*Hbs[j][l]*Hbs[j][m]
                        elseif abs(m-l) == 2
                            Hsc[j,k] += (2,d,C6,R)*Hbs[j][l]*Hbs[j][m]
                        else
                            continue
                        end
                    end
                end
            end
        end
    end
    return Hsc
end

function Htot(;::Float64, J::Float64, g::Float64, d::Float64, C6::Float64, R::Float64)
    Htot = Htit() - J*Hhop + (g/2)*Honsite + (1/2)*Hsc(d,C6,R)
    return Htot
end
```

```julia
J = 1.0
#g = -0.137
g = -0.05
d = 1.5
C6 = 100.0
R = 3.0
#gammas = -5.0:0.1:5.0
gammas = 0:0.2:10
eigs = zeros(Float64, Lb, length(gammas));
```

```julia
H0 = Htot(=0.0, J=J, g=g, d=d, C6=C6, R=R);
H1 = Htot(=1.0, J=J, g=g, d=d, C6=C6, R=R);
H15 = Htot(=1.5, J=J, g=g, d=d, C6=C6, R=R);
H2 = Htot(=2.0, J=J, g=g, d=d, C6=C6, R=R);
H3 = Htot(=3.0, J=J, g=g, d=d, C6=C6, R=R);
H4 = Htot(=4.0, J=J, g=g, d=d, C6=C6, R=R);
H6 = Htot(=6.0, J=J, g=g, d=d, C6=C6, R=R);
H7 = Htot(=7.0, J=J, g=g, d=d, C6=C6, R=R);
H8 = Htot(=8.0, J=J, g=g, d=d, C6=C6, R=R);
H9 = Htot(=9.0, J=J, g=g, d=d, C6=C6, R=R);
H10 = Htot(=10.0, J=J, g=g, d=d, C6=C6, R=R);
```

## Quenching Dynamics

```julia
# I now only makes this function works given the initial state with equal
# number of atoms in each site. But I potentially can extend it to arbitrary initial state.
function evolEE(;H::Matrix{Float64},t::Float64,basis::Vector{Vector{Int64}},init::Vector{Int})
    
    if t == 0.0
        EE = 0.0
    else
        vec = state_evolution(H=H,t=t,basis=basis,init=init)
        #rdmat = reduced_density_matrix(Na=Na, state=vec, Hbs=Hbs)
        rdmat = reduced_density_matrix(state=vec, Na=Na, Ls=Ls, Rd=1)
        vals = svd(rdmat).S
        #EE = -transpose(conj(vals).*vals)*log.(conj(vals).*(vals))
        #vals_squd = vals.*conj(vals)
        EE = -transpose(vals)*log.(vals)
    end
    
    return EE
end
```

```julia
function evolEE_random(;H::Matrix{Float64},t::Float64,basis::Vector{Vector{Int64}},init::Vector{Float64})
    
    if t == 0.0
        rdmat = reduced_density_matrix(state=init, Na=Na, Ls=Ls, Rd=1)
        vals = svd(rdmat).S
        EE = -transpose(vals)*log.(vals)
    else
        vec = state_evolution_random(H=H,t=t,basis=basis,init=init)
        #rdmat = reduced_density_matrix(Na=Na, state=vec, Hbs=Hbs)
        rdmat = reduced_density_matrix(state=vec, Na=Na, Ls=Ls, Rd=1)
        vals = svd(rdmat).S
        #EE = -transpose(conj(vals).*vals)*log.(conj(vals).*(vals))
        #vals_squd = vals.*conj(vals)
        EE = -transpose(vals)*log.(vals)
    end
    
    return EE
end
```

```julia
# Here the quench will happen at time t1.
function quench(;H1::Matrix{Float64}, H2::Matrix{Float64},
        t0::Float64, t1::Float64, t2::Float64,
        basis::Vector{Vector{Int64}}, init::Vector{Int64},
        Na::Int64, Ls::Int64, Rd::Int64)
    
    @assert t0<=t1<=t2
    
    ts_init = t0:0.1:t1
    ts_quenched = t1:0.1:t2
    ts_quenched = ts_quenched[2:end]
    
    init_len = length(ts_init)
    quench_len = length(ts_quenched)
    
    # below we calculate the unquenched EEs
    EEs_unquenched = vec(zeros(Float64, 1, init_len))
    for i in 1:init_len
        init_vec = state_evolution(H=H1, t=ts_init[i], basis=Hbs, init=init)
        init_rdmat = reduced_density_matrix(state=init_vec, Na=Na, Ls=Ls, Rd=Rd)
        init_vals = svd(init_rdmat).S
        init_EE = -transpose(init_vals)*log.(init_vals)
        EEs_unquenched[i] = init_EE
    end
    
    # below we calculate the quenched EEs
    EEs_quenched = vec(zeros(Float64, 1, quench_len))
    init_vec = state_evolution(H=H1, t=t1, basis=Hbs, init=init)
    for i in 1:quench_len
        evol_vec = state_evolution_random(H=H2, t=ts_quenched[i], basis=Hbs, init=init_vec)
        evol_rdmat = reduced_density_matrix(state=evol_vec, Na=Na, Ls=Ls, Rd=1)
        evol_vals = svd(evol_rdmat).S
        evol_EE = -transpose(evol_vals)*log.(evol_vals)
        EEs_quenched[i] = evol_EE
    end
    
    return append!(EEs_unquenched, EEs_quenched)
end
        
```

```julia
@time quench_data = quench(H1=H4, H2=H2, t0=8.0, t1=12.0, t2=16.0, basis=Hbs, init=vec([20 20 20]), Na=Na, Ls=Ls, Rd=1)
```
