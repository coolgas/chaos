## Setting Parameters for BEC

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

## Setting up $a$, $a^\dagger$, and $I$

Matrix representation of the annihilation and creation operators:

$a^\dagger$ = \begin{pmatrix}
0 & 0 & 0 & 0 & \dots & 0 & \dots \\
\sqrt{1} & 0 & 0 & 0 & \dots & 0 & \dots \\
0 & \sqrt{2} & 0 & 0 & \dots & 0 & \dots \\
0 & 0 & \sqrt{3} & 0 & \dots & 0 & \dots \\
\vdots & \vdots & \vdots & \ddots & \ddots & \dots & \dots \\
0 & 0 & 0 & \dots & \sqrt{n} & 0 & \dots & \\
\vdots & \vdots & \vdots & \vdots & \vdots & \ddots & \ddots \end{pmatrix}


$a$ =\begin{pmatrix}
0 & \sqrt{1} & 0 & 0 & \dots & 0 & \dots \\
0 & 0 & \sqrt{2} & 0 & \dots & 0 & \dots \\
0 & 0 & 0 & \sqrt{3} & \dots & 0 & \dots \\
0 & 0 & 0 & 0 & \ddots & \vdots & \dots \\
\vdots & \vdots & \vdots & \vdots & \ddots & \sqrt{n} & \dots \\
0 & 0 & 0 & 0 & \dots & 0 & \ddots \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \ddots \end{pmatrix}

```python
x = range(1, n+1)
x = np.array(x)
x
```

```python
rootx = np.sqrt(x)
rootx
```

```python
a = np.diag(rootx, 1)
adagger = np.transpose(a)
I = np.identity(n+1)
print(a.shape)
print(adagger.shape)
print(I.shape)
```

And the number operator is defined by:
$\hat{N}_i=\hat{a}_i^\dagger \hat{a}_i$

```python
N = adagger @ a
print(N)
```

Calculation with multiple particles deals with particles in different Hilbert spaces, and therefore the creation of the Hamiltonian requires the kronecker delta (tensor product) of the operators

```python
NN = N @ N
NN_I_1 = np.kron(np.kron(NN,I), I)
NN_I_2 = np.kron(I, np.kron(NN,I))
NN_I_3 = np.kron(I, np.kron(I,NN))

N_1_N_2 = np.kron(np.kron(N, N), I)
N_2_N_3 = np.kron(I, np.kron(N, N))
N_1_N_3 = np.kron(N, np.kron(I,N))
print(N_1_N_2.shape)
print(N_2_N_3.shape)
print(N_1_N_3.shape)
print(NN_I_1.shape)
```

```python
adagger_a1 = np.kron(np.kron(adagger, a), I)
adagger_a2 = np.kron(np.kron(a, adagger), I)
adagger_a3 = np.kron(I, np.kron(adagger, a))
adagger_a4 = np.kron(I, np.kron(a, adagger))
print(adagger_a1.shape)
```

```python
N_sing_1 = np.kron(np.kron(N, I), I)
N_sing_3 = np.kron(I, np.kron(I, N))

print(N_sing_1.shape)
print(N_sing_3.shape)
```

## Setting up the Hamiltonian of the system

The Hamiltonian of a quantum system of N bosons in an aligned triple-well system is:
$$\begin{align}
\label{QH}
	\hat{H} =& \frac{U}{N}\left(\hat{N}_1^2-\hat{N}_2^2+\hat{N}_3^2\right)
	+ \epsilon\left(\hat{N}_3-\hat{N}_1\right) \nonumber \\
	&+ \frac{2 U}{N} \left( - \hat{N}_1 \hat{N}_2 - \hat{N}_2 \hat{N}_3 + \hat{N}_1 \hat{N}_3 \right)
	\nonumber \\
	&+\frac{J}{\sqrt{2}}\left(\hat{a}_1^\dagger \hat{a}_2 + \hat{a}_2^\dagger \hat{a}_1\right)+\frac{J}{\sqrt{2}}\left(\hat{a}_2^\dagger \hat{a}_3 + \hat{a}_3^\dagger \hat{a}_2\right),
\end{align}$$

Where $J$ is the tunneling amplitude and $U$ is the onsite interaction strength and strength of interaction between wells

```python
HMTLN = (U/n)*(NN_I_1 - NN_I_2 + NN_I_3) + E*(N_sing_3 - N_sing_1) + (2*U/n)*(-N_1_N_2 - N_2_N_3 + N_1_N_3) + (J/np.sqrt(2))*(adagger_a1 + adagger_a2) + (J/np.sqrt(2))*(adagger_a3 + adagger_a4)
```

## Reduction of Hilbert Space - Method 1

Previous calculations have expanded to Hilbert space of all particle interactions, whereas now it is important to conserve the number of bosons $N = N_1 + N_2 + N_3 $ and reduce the Hilbert space.

This method involves calculating the Hamiltonian and then removing the rows/columns that do not correlate to a conserved number of bosons

```python
N_1 = np.kron(np.kron(N, I), I)
N_2 = np.kron(I, np.kron(N, I))
N_3 = np.kron(I, np.kron(I, N))
```

```python
total_N = N_1 + N_2 + N_3
```

```python
total_N.shape
```

```python
nums = np.diag(total_N.round())
```

```python
total_N[40,40]
```

```python
indices = np.where(nums != n)
```

```python
HMTLN_new = np.delete(np.delete(HMTLN, indices, 0), indices, 1)
HMTLN_new.shape
```

```python
test_array_new = []
U_test_new = np.linspace(0, 2.5, 50)
J1 = 1
for vals in U_test_new:
    HMTLN_2 = (vals/n)*(NN_I_1 - NN_I_2 + NN_I_3) + (2*vals/n)*(-N_1_N_2 - N_2_N_3 + N_1_N_3) + (J1/np.sqrt(2))*(adagger_a1 + adagger_a2) + (J1/np.sqrt(2))*(adagger_a3 + adagger_a4)
    N_1 = np.kron(np.kron(N, I), I)
    N_2 = np.kron(I, np.kron(N, I))
    N_3 = np.kron(I, np.kron(I, N))
    total_N = N_1 + N_2 + N_3
    nums = np.diag(total_N.round())
    indices = np.where(nums != n)    
    HMTLN_test = np.delete(np.delete(HMTLN_2, indices, 0), indices, 1)
    
    eigenvals_new, eigenvecs_new = linalg.eigh(HMTLN_test)
    eigenvals_new = np.sort(eigenvals_new)
    test_array_new.append(eigenvals_new)
    newList_1 = [x / n for x in test_array_new]
    
plt.plot(U_test_new, newList_1)
plt.xlabel('U/J', fontsize = '12')
plt.ylabel('E/(JN)', fontsize = '12')
#plt.ylim((-1, 2.5))
plt.show()
#print(newList)
```

```python
test_array_new_2 = []
E_test_new = np.linspace(0, 2, 200)
J2 = 1
U2 = 0.7
for vals in E_test_new:
    HMTLN_3 = (U2/n)*(NN_I_1 - NN_I_2 + NN_I_3) + vals*(N_sing_3 - N_sing_1) + (2*U2/n)*(-N_1_N_2 - N_2_N_3 + N_1_N_3) + (J2/np.sqrt(2))*(adagger_a1 + adagger_a2) + (J2/np.sqrt(2))*(adagger_a3 + adagger_a4)
    N_1 = np.kron(np.kron(N, I), I)
    N_2 = np.kron(I, np.kron(N, I))
    N_3 = np.kron(I, np.kron(I, N))
    total_N = N_1 + N_2 + N_3
    nums = np.diag(total_N.round())
    indices = np.where(nums != n)    
    HMTLN_test_2 = np.delete(np.delete(HMTLN_3, indices, 0), indices, 1)
    
    eigenvals_3_new, eigenvecs_3_new = linalg.eigh(HMTLN_test_2)     
    eigenvals_3_new = np.sort(eigenvals_3_new)
    test_array_new_2.append(eigenvals_3_new)
    newList_2 = [x / n for x in test_array_new_2]
    
plt.plot(E_test_new, newList_2)
plt.xlabel(r'$\epsilon /J$', fontsize = '12')
plt.ylabel('E/(JN)', fontsize = '12')
plt.ylim((0.3, 0.9))
plt.xlim((1., 2.5))
plt.show()
#print(newList)
```

## Reduction of Hilbert Space - Method 2

This method involves calculating the commutation value of all the combinations that conserve particle number

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
from numpy import linalg

Na = 20
Ls = 3
```

```python
def sums(length, total_sum):
    if length == 1:
        yield (total_sum,)
    else:
        for value in range(total_sum + 1):
            for permutation in sums(length - 1, total_sum - value):
                yield (value,) + permutation

L = list(sums(Ls, Na))
print('total permutations:',len(L))
```

```python
Hbs = L
lH = len(L)
```

```python
empty = np.zeros(shape = (lH, lH))
empty.shape
```

```python
test = np.zeros(shape = (lH, lH))
#test
```

```python
##Non Universal - checking error - only for triple well
J = 1
U = 0.2

def ApAm(m, Hb, Na, lH, array):
    for j in range(0,lH):
        if Hb[j][m] < Na and Hb[j][m+1] > 0:
            a = Hb[j][m] + 1
            b = Hb[j][m+1] - 1
            if m == 0:
                new_ind = Hb.index(((a), (b), (Na - (a+b))))
            elif m == 1:
                new_ind = Hb.index(((Na - (a+b)), (a), (b)))
            array[j][new_ind] += (J/np.sqrt(2)) * np.sqrt((Hb[j][m]+1)*(Hb[j][m+1]))
```

```python
#test2 = np.zeros(shape = (lH, lH))
U0 = 0.7
m = 0
L_end = 3
def Honsite(Hb, U, arr, lH, Na):
    for j in range(0, lH):
        for k in range(0, lH):
            if j == k:
                for i in range(m, L_end):
                    if i == 1:
                        full_sum = -(U/Na) * (Hb[j][i]**2)
                    else:
                        full_sum = (U/Na) * (Hb[j][i]**2)
                    arr[j][k] += full_sum
```

```python
#test3 = np.zeros(shape = (lH, lH))
U1 = 1.6666666666666667  
m = 0
L_end = 3
def HNN(Hb, U, arr, lH, Na):
    for j in range(0, lH):
        for k in range(0, lH):
            if j == k:
                for i in range(m, L_end-1):
                    full_sum_2 = -(2*U/Na) * (Hb[j][i]) * (Hb[j][i+1])
                    arr[j][k] += full_sum_2
```

```python
HNN(Hbs, U0, HHop, lH, Na)
```

```python
#test4 = np.zeros(shape = (lH, lH))
U2 = 1.6666666666666667  
m = 0
L_end = 3
def HNNN(Hb, U, arr, lH, Na):
    for j in range(0, lH):
        for k in range(0, lH):
            if j == k:
                for i in range(m, L_end-2):
                    full_sum_3 = (2*U/Na) * (Hb[j][i]) * (Hb[j][i+2])
                    arr[j][k] += full_sum_3
```

```python
HNNN(Hbs, U0, HHop, lH, Na)
```

```python
E_test = 1
m = 0
L_end = 3
def Hsing(Hb, E, arr, lH, Na):
    for j in range(0, lH):
        for k in range(0, lH):
            if j == k:
                for i in range(m, L_end):
                    if i == 0:
                        full_sum_4 = -(E) * (Hb[j][i])
                    elif i == 2:
                        full_sum_4 = (E) * (Hb[j][i])
                    else:
                        full_sum_4 = 0
                    arr[j][k] += full_sum_4
```

```python
n_lev = 4
x_1 = range(1, n_lev+1)
x_1 = np.array(x_1)
rootx_1 = np.sqrt(x_1)
a_1 = np.diag(rootx_1, 1)
adagger_1 = np.transpose(a_1)
I_1 = np.identity(n_lev+1)
N_1 = adagger_1 @ a_1
NN_1 = N_1 @ N_1
print(NN_1.shape)
NN_I_1_1 = np.kron(np.kron(NN_1,I_1), I_1)
print(NN_I_1_1.shape)
NN_I_2_1 = np.kron(I_1, np.kron(NN_1,I_1))
print(NN_I_2_1.shape)
NN_I_3_1 = np.kron(I_1, np.kron(I_1,NN_1))
print(NN_I_3_1.shape)


N_1_N_2_1 = np.kron(np.kron(N_1, N_1), I_1)
print(N_1_N_2_1.shape)
N_2_N_3_1 = np.kron(I_1, np.kron(N_1, N_1))
print(N_2_N_3_1.shape)
N_1_N_3_1 = np.kron(N_1, np.kron(I_1,N_1))
print(N_1_N_3_1.shape)
adagger_a1_1 = np.kron(np.kron(adagger_1, a_1), I_1)
print(adagger_a1_1.shape)
adagger_a2_1 = np.kron(np.kron(a_1, adagger_1), I_1)
print(adagger_a2_1.shape)
adagger_a3_1 = np.kron(I_1, np.kron(adagger_1, a_1))
print(adagger_a3_1.shape)
adagger_a4_1 = np.kron(I_1, np.kron(a_1, adagger_1))
print(adagger_a4_1.shape)
N_sing_1_1 = np.kron(np.kron(N_1, I_1), I_1)
print(N_sing_1_1.shape)
N_sing_3_1 = np.kron(I_1, np.kron(I_1, N_1))
print(N_sing_3_1.shape)
```

```python
diff_list = []
U_1 = 0.7
E_1 = 0
J1 = 1

HMTLN_5 = (U_1/n_lev)*(NN_I_1_1 - NN_I_2_1 + NN_I_3_1) + E_1*(N_sing_3_1 - N_sing_1_1) + (2*U_1/n_lev)*(-N_1_N_2_1 - N_2_N_3_1 + N_1_N_3_1) + (J1/np.sqrt(2))*(adagger_a1_1 + adagger_a2_1) + (J1/np.sqrt(2))*(adagger_a3_1 + adagger_a4_1)
N_1_1 = np.kron(np.kron(N_1, I_1), I_1)
N_2_1 = np.kron(I_1, np.kron(N_1, I_1))
N_3_1 = np.kron(I_1, np.kron(I_1, N_1))
total_N = N_1_1 + N_2_1 + N_3_1
nums = np.diag(total_N.round())
indices = np.where(nums != n)    
HMTLN_del = np.delete(np.delete(HMTLN_5, indices, 0), indices, 1)
    
eigenvals_new, eigenvecs_new = linalg.eig(HMTLN_del)
eigenvals_new = np.sort(eigenvals_new)

for i in range(1,len(eigenvals_new)):
    x = eigenvals_new[i] - eigenvals_new[i-1]
    diff_list.append(x)
```
