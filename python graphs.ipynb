### Fitting Brody distributions

```python
import os

from scipy.special import gamma
from scipy.optimize import curve_fit
```

```python
x = np.arange(1.0,8.0,0.1)
lambdas = np.arange(2.0,15.05,0.05)
```

```python
def model(x,p):
    b = gamma((p+2)/(p+1))**(p+1)
    return (p+1)*b*x**(p)*np.exp(-b*x**(p+1))
```

```python
fits = []
for r, d, f in os.walk('./new_data/gamma=2.5/N=60/'):
    for file in sorted(f):
        if '.txt' in file:
            normNenergy = []
            path = os.path.join(r, file)
            with open(path) as f:
                lines = f.readlines()
                for i in lines:
                    normNenergy.append(float(i[:-2]))
                fits.append(curve_fit(model, x, normNenergy, p0=1.0)[0][0])
                
fits2 = []
for r, d, f in os.walk('./new_data/gamma=2.5/N=90/'):
    for file in sorted(f):
        if '.txt' in file:
            normNenergy = []
            path = os.path.join(r, file)
            with open(path) as f:
                lines = f.readlines()
                for i in lines:
                    normNenergy.append(float(i[:-2]))
                fits2.append(curve_fit(model, x, normNenergy, p0=1.0)[0][0])
                
fits3 = []
for r, d, f in os.walk('./new_data/gamma=2.5/N=120/'):
    for file in sorted(f):
        if '.txt' in file:
            normNenergy = []
            path = os.path.join(r, file)
            with open(path) as f:
                lines = f.readlines()
                for i in lines:
                    normNenergy.append(float(i[:-2]))
                fits3.append(curve_fit(model, x, normNenergy, p0=1.0)[0][0])
```

```python
fig4, ax4 = plt.subplots(1,1, figsize=(20, 8))
ax4.plot(lambdas, fits, marker='^',color='red', linewidth=4, markersize=10, label=r"$N=60$")
ax4.plot(lambdas, fits2, marker='D', linewidth=4, markersize=10, label=r"$N=90$")
ax4.plot(lambdas, fits3, marker='s', linewidth=4, markersize=10, label=r"$N=120$")
ax4.set_xlim(xmin=2, xmax=15)
ax4.set_xlabel(r"$U$",fontsize=45)
ax4.set_ylabel(r"$\beta$",fontsize=45)
ax4.legend(fontsize = 'x-small')
plt.savefig("./pics/pypic4.png", format='png', bbox_inches='tight',pad_inches = 0.4)
```
