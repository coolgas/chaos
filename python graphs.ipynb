## Average values of ee and std

```python
ave = []
ave_new = []
ave_newnew = []

with open("./data/aves/avee/aveEEs.txt", "r") as fin:
    for d in fin:
        ave.append(float(d[:-2]))

with open("./data/aves/avee_new/aveEEs_new.txt", "r") as fin:
    for d in fin:
        ave_new.append(float(d[:-2]))

with open("./data/aves/avee_newnew/aveEEs_newnew.txt", "r") as fin:
    for d in fin:
        ave_newnew.append(float(d[:-2]))
        
        
std = []
std_new = []
std_newnew = []

with open("./data/aves/log_stdee/text0.txt", "r") as fin:
    for d in fin:
        std.append(float(d[:-2]))
        
with open("./data/aves/log_stdee_new/text0.txt", "r") as fin:
    for d in fin:
        std_new.append(float(d[:-2]))

with open("./data/aves/log_stdee_newnew/text0.txt", "r") as fin:
    for d in fin:
        std_newnew.append(float(d[:-2]))
```

```python
fig, ax = plt.subplots(3,2, figsize=(33, 33))

ax[0,0].plot(ts, evolEE0s, color='red', linestyle='--', linewidth=6, label=r"$\gamma=0$")
ax[0,0].plot(ts, evolEE4s, color='blue', linestyle="dotted", linewidth=6, label=r"$\gamma=4$")
ax[0,0].plot(ts, evolEE8s, color='black', linewidth=6, label=r"$\gamma=8$")
ax[0,0].set(ylabel=r"$S_{EE}$")
ax[0,0].legend()
ax[0,0].annotate("(a)",xy=(19.1,3.7))

ax[1,0].plot(ts, evolEE0s_new, color='red', linestyle='--', linewidth=6)
ax[1,0].plot(ts, evolEE4s_new, color='blue', linestyle="dotted", linewidth=6)
ax[1,0].plot(ts, evolEE8s_new, color='black', linewidth=6)
ax[1,0].set(ylabel=r"$S_{EE}$")
ax[1,0].annotate("(b)",xy=(19.1,3.9))

ax[2,0].plot(ts, evolEE0s_newnew, color='red', linestyle='--', linewidth=6)
ax[2,0].plot(ts, evolEE4s_newnew, color='blue', linestyle="dotted", linewidth=6)
ax[2,0].plot(ts, evolEE8s_newnew, color='black', linewidth=6)
ax[2,0].set(ylabel=r"$S_{EE}$")
ax[2,0].set(xlabel=r'$tJ$')
ax[2,0].annotate("(c)", xy=(19.2,3.8))

ax[0,1].scatter(gamma_new, np.log(ave),s=300)
ax[0,1].errorbar(gamma_new, np.log(ave), std, fmt='o', capsize=12, elinewidth=3)
ax[0,1].set(ylabel=r"$log(\bar{S}_{EE})$")
ax[0,1].annotate("(d)", xy=(7.65,1.36))

ax[1,1].scatter(gamma_new, np.log(ave_new), s=300)
ax[1,1].errorbar(gamma_new, np.log(ave_new), std_new, fmt='o', capsize=12, elinewidth=3)
ax[1,1].set(ylabel=r"$log(\bar{S}_{EE})$")
ax[1,1].annotate("(e)", xy=(7.65,1.2))

ax[2,1].scatter(gamma_new, np.log(ave_newnew), s=300)
ax[2,1].errorbar(gamma_new, np.log(ave_newnew), std_newnew, fmt='o', capsize=12, elinewidth=3)
ax[2,1].set(ylabel=r"$log(\bar{S}_{EE})$")
ax[2,1].set(xlabel=r"$\gamma/J$")
ax[2,1].annotate("(f)", xy=(7.75,1.38))

plt.savefig("./pics/pypic1.png", format='png', bbox_inches='tight',pad_inches = 0.4)
```
